```python
# Importing Libraries
import ast
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt  

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df["job_skills"].apply(lambda Skill_list: ast.literal_eval(Skill_list) if pd.notna(Skill_list) else Skill_list)
```

```python
df_us = df[(df['job_title_short']== "Data Analyst") & (df['job_country']=="United States")].copy()
df_us 
```

```python
df_us = df_us.dropna(subset=['salary_year_avg']) #to plot boxplot dropna value
df_us
```

```python
df_us["salary_year_avg"].head(10)
```

```python
df_us["salary_year_avg"].plot(kind = 'box', vert=False)
```

```python
#comparing box plots of different job titles
job_titles = ["Data Analyst", "Data Scientist", "Data Engineer"]
```

```python
df_us_three = df[(df['job_title_short'].isin (job_titles) ) & (df['job_country']=="United States")].copy()
df_us_three.job_title_short.value_counts()
```

```python
df_us_three = df_us_three.dropna(subset=['salary_year_avg']) #to plot boxplot dropna value
df_us_three
```

```python
title_jobslist =[df_us_three[df_us_three["job_title_short"]== job_title ]["salary_year_avg"] for job_title in job_titles]
title_jobslist
```

```python
plt.boxplot(title_jobslist, labels=job_titles, vert=False)
plt.xlabel('Average Yearly Salary')
plt.title('Distribution of Average Yearly Salary for Different Data Job Titles in the US')
ax = plt.gca()
ax.xaxis.set_major_formatter(plt.FuncFormatter(lambda x , pos: f"${x/1000}k")) # Format y-axis labels as $Xks
```

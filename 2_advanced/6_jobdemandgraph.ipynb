```python
# Importing Libraries
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt  

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
job_usa = df[df["job_country"]=="United States"].copy()
job_usa
```

```python
id(job_usa) == id(df) # False, because job_usa is a new object in memory
```

```python
job_usa["job_month"] = job_usa["job_posted_date"].dt.strftime('%B') # Full month name
job_usa
```

```python
pivot_usa = job_usa.pivot_table(index='job_month', columns='job_title_short', aggfunc='size') 
pivot_usa # DataFrame with months as index and job titles as columns
```

```python
pivot_usa.reset_index(inplace=True)
pivot_usa 
```

```python
pivot_usa["month_number"] = pd.to_datetime(pivot_usa["job_month"], format='%B').dt.month
pivot_usa["month_number"] # Numeric representation of the month
```

```python
pivot_usa 
```

```python
pivot_usa.sort_values("month_number", inplace=True)
pivot_usa # Sorted by month number
```

```python

pivot_usa 
```

```python
pivot_usa.drop("month_number" , axis=1, inplace=True) 
pivot_usa
```

```python
pivot_usa 
```

```python
pivot_usa.set_index('job_month' , inplace=True)
pivot_usa 
```

```python
pivot_usa.plot(kind='line')
```

```python
top_usa_job =job_usa.value_counts('job_title_short').head(3).index
```

```python
pivot_usa_top = pivot_usa[top_usa_job]
pivot_usa_top
```

```python
pivot_usa_top.plot(kind='line') 
plt.title('Top 3 Job Demand in USA')
plt.ylabel('Number of Job Posts')
plt.xlabel('Month') 
```

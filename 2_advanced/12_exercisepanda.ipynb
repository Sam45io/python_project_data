```python
# Importing Libraries
import ast
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt  

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df["job_skills"].apply(lambda Skill_list: ast.literal_eval(Skill_list) if pd.notna(Skill_list) else Skill_list)
```

```python
df_data = df[df['job_title_short']== 'Data Analyst'].copy()
df_data
```

```python
df_data["Job_posted_month_no"] = df_data["job_posted_date"].dt.month
df_data 
```

```python
df_data_explode = df_data.explode('job_skills')
df_data_explode
```

```python
pivot_explode =df_data_explode.pivot_table(index='Job_posted_month_no', columns='job_skills', aggfunc='size', fill_value=0)
pivot_explode
```

```python
pivot_explode.loc["total"] = pivot_explode.sum()
pivot_explode
```

```python
pivot_explode.loc["total"].sort_values(ascending=False).index # Sorting the columns based on total values
```

```python
pivot_explode[pivot_explode.loc["total"].sort_values(ascending=False).index]
```

```python
df_da_pivot = pivot_explode[pivot_explode.loc["total"].sort_values(ascending=False).index]
df_da_pivot.drop("total", inplace=True)
```

```python
df_da_pivot
```

```python
df_da_pivot.iloc[:,:5] 
```

```python
df_da_pivot.iloc[:,:5].plot(kind='line')
plt.title('Top 5 Skills Required for Data Analyst Role Over Months')
plt.xlabel('Month Number')
plt.ylabel('Number of Job Posts')
plt.legend(title='Skills')
```

```python
# Importing Libraries
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt  

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df_US = df[df['job_country'] == 'United States'].copy()
df_US['job_posted_month'] = df_US['job_posted_date'].dt.strftime('%B')
df_US_pivot = df_US.pivot_table(index='job_posted_month', columns='job_title_short', aggfunc='size')
df_US_pivot = df_US_pivot.reset_index()
df_US_pivot['job_posted_month_no'] = pd.to_datetime(df_US_pivot['job_posted_month'], format='%B').dt.month
df_US_pivot = df_US_pivot.sort_values('job_posted_month_no')
df_US_pivot = df_US_pivot.set_index('job_posted_month')
df_US_pivot = df_US_pivot.drop(columns='job_posted_month_no')

df_US_pivot
```

```python
us_soft_pivot = pd.read_csv(r"C:\Users\sampu\Downloads/software_jobs.csv", index_col='job_posted_month')
us_soft_pivot 
```

```python
merged_data_us = df_US_pivot.merge(us_soft_pivot, on ='job_posted_month') 
merged_data_us
```

```python
top_5jobs = merged_data_us.sum().sort_values(ascending=False).head().index # Get the top 5 job titles
top_5jobs 
```

```python
merged_data_us[top_5jobs].plot(kind="line") # Plotting the top 5 job trends
plt.title("Top 5 Job Trends in the US")
plt.xlabel("Month") 
plt.ylabel("Number of Job Postings")    
plt.legend(title="Job Titles") 
plt.ylim(0,20000)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
```

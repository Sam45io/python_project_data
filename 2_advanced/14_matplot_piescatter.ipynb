```python
# Importing Libraries
import ast
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt  

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df["job_skills"].apply(lambda Skill_list: ast.literal_eval(Skill_list) if pd.notna(Skill_list) else Skill_list)
```

```python
df["job_work_from_home"].value_counts().plot(kind='pie', startangle=90 , autopct='%1.1f%%')
plt.title("Job Work From Home Distribution")
plt.ylabel("")
```

```python
df["job_title_short"].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title("Job Title Distribution")
plt.ylabel("")
plt.tight_layout()
```

```python
df[["job_work_from_home","job_no_degree_mention","job_health_insurance"]]
```

```python
fig, ax = plt.subplots(1,3)
dict_coloumn = {
    "job_work_from_home": "Work From Home",
    "job_no_degree_mention": "No Degree Mention",
    "job_health_insurance": "Health Insurance"
    }
for i ,(coloumn, title) in enumerate(dict_coloumn.items()):
    ax[i].pie(df[coloumn].value_counts(), startangle=90 , autopct='%1.1f%%' , labels = ['False' , 'true'])
    ax[i].set_title(title)
    ax[i].set_xlabel("")
    ax[i].set_ylabel(" ")
```

```python
df_jobda = df[df['job_title_short'] == 'Data Analyst'].copy()
df_jobda
```

```python
df_exploded = df.explode('job_skills')
df_exploded
```

```python
skill_countss = df_exploded.groupby('job_skills').agg(
    skill_count = ('job_skills','count'),
    median_salary = ('salary_year_avg','median')
)
skill_countss
```

```python
plot_skill = skill_countss.sort_values(by='skill_count', ascending = False).head(10)
plot_skill 
```

```python
plot_skill.plot(kind="scatter", x='skill_count',y='median_salary')
plt.title("Top 10 skill : job vs median salary")
plt.xlabel("Job Count")
plt.ylabel("Median Salary")
for i , txt in enumerate(plot_skill.index):
    plt.annotate(txt, (plot_skill['skill_count'][i], plot_skill['median_salary'][i]))

    ax = plt.gca()
    ax.yaxis.set_major_formatter(plt.FuncFormatter(lambda y , pos: f"${y/1000}k")) # Format y-axis labels as $Xks

```

```python
# Importing Libraries
import ast
import pandas as pd
import seaborn as sns
from datasets import load_dataset
import matplotlib.pyplot as plt  

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df["job_skills"].apply(lambda Skill_list: ast.literal_eval(Skill_list) if pd.notna(Skill_list) else Skill_list)
```

```python
df_us = df[(df['job_title_short']== "Data Analyst") & (df['job_country']=="United States")].copy()
df_us 
```

```python
df_us["Job_posted_month_no"] = df_us["job_posted_date"].dt.month
df_us
```

```python
df_us_explode = df_us.explode("job_skills")
df_us_explode
```

```python
df_us_pivot =df_us_explode.pivot_table(index="Job_posted_month_no",columns="job_skills",aggfunc="size",fill_value=0)
df_us_pivot
```

```python
df_us_pivot.loc["Total"]=df_us_pivot.sum() 
df_us_pivot
```

```python
df_us_pivot =df_us_pivot[df_us_pivot.loc["Total"].sort_values(ascending=False).index]
df_us_pivot
```

```python
df_us_pivot = df_us_pivot.drop("Total")
df_us_pivot
```

```python
df_us_total = df_us.groupby("Job_posted_month_no").size() #total job posting in a month
df_us_total
```

```python
df_us_percent=df_us_pivot.div(df_us_total/100,axis=0)
df_us_percent
```

```python
# changes month number to month name
df_us_percent = df_us_percent.reset_index()
df_us_percent['Job_posted_month'] = df_us_percent['Job_posted_month_no'].apply(lambda x: pd.to_datetime(x, format='%m').strftime('%b'))
df_us_percent = df_us_percent.set_index('Job_posted_month')
df_us_percent = df_us_percent.drop(columns='Job_posted_month_no')
df_us_percent
```

```python
df_plot = df_us_percent.iloc[:,:5]
df_plot
```

```python
sns.lineplot(data=df_plot,dashes=False,palette="tab10")
sns.set_theme(style="ticks")
sns.despine()
plt.title('Top 5 Skills Required for Data Analyst Role Over Months')
plt.xlabel('Month Number')
plt.ylabel('liklihood of Job Posts')
plt.legend(title='Skills')
plt.tight_layout()
plt.legend().remove()

from matplotlib.ticker import PercentFormatter
ax=plt.gca()
ax.yaxis.set_major_formatter(PercentFormatter(decimals=0))
for i in range(5):
    plt.text(11.2,df_plot.iloc[-1,i],df_plot.columns[i])
   
```

```python
# Importing Libraries
import ast
import pandas as pd
import seaborn as sns
from datasets import load_dataset
import matplotlib.pyplot as plt  

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df["job_skills"].apply(lambda Skill_list: ast.literal_eval(Skill_list) if pd.notna(Skill_list) else Skill_list)
```

```python
df_us = df[df["job_country"]=="United States"]
df_us
```

```python
df_skills = df_us.explode("job_skills")
df_skills
```

```python
df_skill_count = df_skills.groupby(["job_skills","job_title_short"]).size()
df_skill_count
```

```python
df_skill_count = df_skill_count.reset_index(name="skill_count")
df_skill_count
```

```python
df_skill_count.sort_values(by="skill_count",ascending=False,inplace=True)
df_skill_count
```

```python
job_titles = df_skill_count["job_title_short"].unique().tolist() #array to list
job_titles
```

```python
job_titles = sorted(job_titles[:3])
job_titles
```

```python
fig, ax = plt.subplots(len(job_titles),1)
#fig, ax = plt.subplots(3,1)

for i, title in enumerate(job_titles): # Looping through the job titles
    df_title = df_skill_count[df_skill_count['job_title_short'] == title].head(5) # Filtering the DataFrame for the current job title and getting the top 5 skills
    df_title.plot(kind='barh', x='job_skills', y='skill_count', ax=ax[i], title=title) # Plotting a horizontal bar chart
    ax[i].invert_yaxis()  # Invert y-axis to have the highest count on top
    ax[i].set_ylabel('')  # Removing y-axis label for cleaner look
    ax[i].legend().remove()  # Removing legend for cleaner look
fig.suptitle('Top 5 Skills by Job Title', fontsize=16)  # Adding a main title for the figure
fig.tight_layout()  # Adjust layout to prevent overlap
```

```python
# % of skill in job posting = (skill count in job posting/total count of job posting)
df_job_title_count=df_us["job_title_short"].value_counts().reset_index(name="jobs_total")
df_job_title_count
```

```python
df_skill_percent =pd.merge(df_skill_count,df_job_title_count,how="left",on="job_title_short")
df_skill_percent["skill_percentage"]=100*df_skill_percent["skill_count"]/df_skill_percent["jobs_total"]
df_skill_percent
```

```python
fig, ax = plt.subplots(len(job_titles),1)
#fig, ax = plt.subplots(3,1)
sns.set_theme(style="ticks")

for i, title in enumerate(job_titles): # Looping through the job titles
    df_title = df_skill_percent[df_skill_percent['job_title_short'] == title].head(5) # Filtering the DataFrame for the current job title and getting the top 5 skills
    #df_title.plot(kind='barh', x='job_skills', y='skill_percentage', ax=ax[i], title=title) # Plotting a horizontal bar chart
    sns.barplot(data=df_title,x="skill_percentage",y="job_skills",ax=ax[i],hue="skill_count",palette="dark:b_r")
    #ax[i].invert_yaxis()  # Invert y-axis to have the highest count on top
    ax[i].set_ylabel('')  # Removing y-axis label for cleaner look
    ax[i].set_xlabel('')
    ax[i].legend().remove()  # Removing legend for cleaner look
    ax[i].set_title(title)
    ax[i].set_xlim(0,78)
    for n,v in enumerate(df_title["skill_percentage"]):
        ax[i].text(v+1,n,f"{v:.0f}%",va="center")
        if i != len(job_titles) -1:
         ax[i].set_xticks([])

fig.suptitle('Liklihood of us job posting', fontsize=16)  # Adding a main title for the figure
fig.tight_layout()  # Adjust layout to prevent overlap
```
